```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

# Gerando dados de exemplo (substitua isso com seus dados)
# Vamos criar uma sequência de 1000 valores e uma janela de 10 como exemplo
data = np.sin(np.linspace(0, 100, 1000))  # Exemplo de uma série temporal senoidal
X, y = [], []
window_size = 10
for i in range(len(data) - window_size):
    X.append(data[i:i + window_size])
    y.append(data[i + window_size])

# Convertendo para numpy arrays e remodelando para (n amostras, janela de tempo, características)
X = np.array(X)
y = np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))  # (n amostras, janela de tempo, 1 característica)




# Construindo o modelo LSTM
model = Sequential()
model.add(LSTM(50, activation='tanh', input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Treinando o modelo
model.fit(X, y, epochs=20, verbose=1)

# Fazendo previsões (usando os primeiros 10 valores da série)
prediction = model.predict(np.array([data[:window_size]]).reshape(1, window_size, 1))
print("Previsão para o próximo valor:", prediction[0][0])
```

```python
import numpy as np
import pandas as pd

def create_crypto_dataset(N, w, rows=2920):
    """
    Cria um dataset sintético para simular uma série temporal de retornos logarítmicos de criptomoedas.
    
    Parâmetros:
    N (int): Número de criptomoedas.
    w (int): Número de dias passados (janelas temporais).
    rows (int): Número de linhas (dias), default é 2920 (aproximadamente 8 anos de dados diários).
    
    Retorna:
    pd.DataFrame: Dataset com colunas de atributos e uma coluna alvo.
    """
    # Inicializando os dados aleatórios para os retornos logarítmicos
    np.random.seed(42)
    
    # Criando dados sintéticos para retornos logarítmicos (N criptomoedas, w dias anteriores)
    X = np.random.randn(rows, N * w)  # (linhas, N*w atributos)
    
    # Idade da criptomoeda em dias (variando entre 0 e o número de linhas)
    age = np.random.randint(1, rows + 1, size=(rows, N))
    
    # Concatenando retornos logarítmicos e a idade da criptomoeda
    X_with_age = np.concatenate([X, age], axis=1)
    
    # Criando os retornos logarítmicos do dia (coluna alvo)
    y = np.random.randn(rows, N)
    
    # Convertendo para DataFrame com nomes de colunas
    column_names = [f'crypto_{i+1}_return_{j+1}' for i in range(N) for j in range(w)]
    column_names += [f'crypto_{i+1}_age' for i in range(N)]
    column_names += [f'crypto_{i+1}_target' for i in range(N)]
    
    # Criando o DataFrame com atributos e alvos
    dataset = pd.DataFrame(np.concatenate([X_with_age, y], axis=1), columns=column_names)
    
    return dataset

# Exemplo de criação de dataset com N=3 criptomoedas e w=5 dias
N = 20  # número de criptomoedas
w = 60  # número de dias passados
dataset = create_crypto_dataset(N, w)

# Mostrar as primeiras linhas do dataset
dataset.head()
```
